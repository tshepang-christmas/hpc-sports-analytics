```python
import pandas as pd

df = pd.read_csv('hpc_football_data.csv', index_col=0)
df
```

```python
position_avg = df.groupby('position')['vertical_jump_cm'].mean().sort_values()
print(position_avg)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.bar(position_avg.index, position_avg.values)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(position_avg.index, position_avg.values, color='steelblue')
plt.xlabel('Position')
plt.ylabel('Average Vertical Jump (cm)')
plt.title('Average Vertical Jump by Football Position')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['height_cm'], df['vertical_jump_cm'])
plt.xlabel('Height (cm)')
plt.ylabel('Vertical Jump (cm)')
plt.title('Height vs Vertical Jump Performance')
plt.show()
```
